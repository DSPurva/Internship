```python
# scrape all product urls
product_urls = []
start=0
end=3

for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]')     
    nxt_button.click()
    time.sleep(2)
    
# MAking empty lists
Brand = []
Name = []
Price=[]
Return_Exchange=[]
Expected_delivery=[]
Availability=[]


for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
# Extracting Brand from xpath
    
    try:
        brand = driver.find_element(By.XPATH,'//a[@id="bylineInfo"]')
        Brand.append(brand.text)

    except NoSuchElementException:
        Brand.append('-')
    
        
# Extracting Name from xpath        
    
    
    try:
        n= driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        Name.append(n.text)

    except NoSuchElementException:
        Name.append('-')
    

# Extracting Price from xpath
    
    try:
        p=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(p.text)

    except NoSuchElementException:
        Price.append('-')
    
# Extracting Return_Exchange from xpath
    
    try:
        re=driver.find_element(By.XPATH,'//a[@class="a-size-small a-link-normal a-text-normal"]')
        Return_Exchange.append(re.text)

    except NoSuchElementException:
        Return_Exchange.append('-')



    # Extracting Expected_delivery from xpath
    
    try:
        ed=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[1]/div[1]/div[2]/div[2]/div/div/div[2]/div[3]/div/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[8]/div[1]/div/div/div[2]/span/span[1]')
        Expected_delivery.append(ed.text)

    except NoSuchElementException:
        Expected_delivery.append('-')

    # Extracting Availability from xpath    
    
    try:
        a= driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(a.text)

    except NoSuchElementException:   
        Availability.append('-')

```

```python
#creating empty lists

hostel_name=[]
distance=[]
total_review=[]
overall_review=[]
price=[]
facility=[]
Rating=[]
```

```python

while(True):
    
    # Hostel name
    names = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for name in names:
        hostel_name.append(name.text)
    time.sleep(4)
        
    # Distance from city centre
    dis = driver.find_elements(By.XPATH,'//span[@class="description"]')
    for d in dis:
        distance.append(d.text)
    time.sleep(4)
        
    # Overall Review    
    review = driver.find_elements(By.XPATH,'//div[@class="keyword"]//span')
    for r in review:
        overall_review.append(r.text)
    time.sleep(4)
    
    # Total No of reviews     
    t_review = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for tr in t_review:
        total_review.append(tr.text)
    time.sleep(4)
    
    # facilities
    service = driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for s in service:
        facility.append(s.text)
    time.sleep(4)
    
    # Prices    
    prices = driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    for p in prices:
        price.append(p.text)
    time.sleep(4)    
        
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
        
time.sleep(4)        
        
# Separate  Privates_From price  and  Dorms_From price
Privates_From = []
for i in range(0,len(price),2):
    Privates_From.append(price[i])
time.sleep(4)

Dorms_From = []
for i in range(1,len(price),2):
    Dorms_From.append(price[i])  
```

```python
# scrape Hostels URL
hostel_url = []

while(True):
    urls = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')
    for url in urls:
        hostel_url.append(url.get_attribute("href"))
    time.sleep(4)    
        
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break


        
Rate = []
descp=[]
for page in hostel_url:
    driver.get(page)
    
    # Rating
    try:
        ratings = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[1]/div[1]/div[1]')
        Rate.append(ratings.text)
    except NoSuchElementException:
        Rate.append("No Rating")  
    time.sleep(4)
    
    
    # Property Description
    try:
        pd = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[2]/div[2]/div/div[2]')
        descp.append(pd.text)
    except NoSuchElementException:
        descp.append("No Description")  

    
time.sleep(2)        
# remove extra data from Rating     
all_text = []
for i in Rate:
    all_text.append(i.split())
time.sleep(4)

for i in all_text:
    Rating.append(i[0]) 
```
