```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#getting web driver and url
driver = webdriver.Chrome(r"D:\Fliprobo\chromedriver.exe")
driver.get("https://github.com/")
```

```python
#selecting explore from the given option
select_explore=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/button")
select_explore.click()
```

```python
#selecting trending from dropdown
select_trending=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a")
select_trending.click()
```

```python
#scrping repository title

title=[]
try:
    t= driver.find_elements(By.XPATH,'//article[@class="Box-row"]/h1/a')
    for i in t:
        title.append(i.text)
except:
    title.append('-')
    

#scraping repository description

desp=[]
try:
    d = driver.find_elements(By.XPATH,'//article[@class="Box-row"]/p')
    for i in d:
        desp.append(i.text)
except:
    desp.append('-')
    
    
## Scrapping language

language=[]
try:
    lang = driver.find_elements(By.XPATH,'//article[@class="Box-row"]//div[2]//span//span[2]')
    for i in lang:
        language.append(i.text)
except:
    language.append('-')


```

```python
print(len(title),len(desp),len(language))
```

```python
# Scrapping urls to get contributor's count:    
urls=[]
match_page=driver.find_elements(By.XPATH,'//article[@class="Box-row"]//h1//a')
for i in match_page:
    urls.append(i.get_attribute('href'))
       
counts=[]
for i in urls:
    driver.get(i)
    time.sleep(2)
    try:
        count =driver.find_element(By.XPATH,'//div[@class="BorderGrid BorderGrid--spacious"]//div[4]//span')
        counts.append(count.text)
    except NoSuchElementException:
        counts.append('N/A')

print(len(counts))
    
```

```python
#creating Dataframe
Repositories = pd.DataFrame({"Repository_Title":title[0:23],
                             "Description":desp[0:23],"Language_used":language,
                             "Contributors_count":counts[0:23]})
```

```python
#getting driver and url
driver = webdriver.Chrome(r"D:\Fliprobo\chromedriver.exe")
driver.get("https://archive.ics.uci.edu/")
```

```python
# clicking on view all dataset

view_all_dataset = driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b').click()
```

```python
#Scraping dataset name
name=[]

try:
    n=driver.find_elements(By.XPATH,"//p[@class='normal']/b/a")
    for i in n:
        name.append(i.text)
except:
    name.append('-')

#scraping dataset types
types=[]
try:
    t=driver.find_elements(By.XPATH,"//table/tbody/tr/td[2]/p[@class='normal']")
    for i in t[0:1245:2]:
        types.append(i.text)
except:
    types.append('-')
types=types[1:]    

#scraping tasks of data    
task=[]
try:
    tasks=driver.find_elements(By.XPATH,'//td[3]//p[@class="normal"]')
    for i in tasks[0:622]:
        task.append(i.text)
except:
    task.append('-')


#scraping attributes of dataset
attribute=[]
try:
    at=driver.find_elements(By.XPATH,'//td[4]//p[@class="normal"]')
    for i in at:
        attribute.append(i.text)
except:
    attribute.append('-')
    
#Scraping no of instances
    
instance=[]
try:
    noi=driver.find_elements(By.XPATH,'//td[5]//p[@class="normal"]')
    for i in noi:
        instance.append(i.text)
except:
    instance.append('-')
    

#scraping no of attributes    
No_attribute=[]
try:
    noa=driver.find_elements(By.XPATH,'//td[6]//p[@class="normal"]')
    for i in noa:
        No_attribute.append(i.text)
except:
    No_attribute.append('-')
    
#scraping year
year=[]
try:
    y=driver.find_elements(By.XPATH,'//td[7]//p[@class="normal"]')
    for i in y:
        year.append(i.text)
except:
    year.append('-')
    
```

```python
print(len(name),len(types),len(task),len(attribute),len(instance),len(No_attribute),len(year))
```

```python
Datasets = pd.DataFrame({"Dataset_name":name,
                        "Dataset_types":types,
                        "Default_task":task,
                        "Attribute_type":attribute,
                        "Instance":instance,
                        "Attribute":No_attribute,
                        "Year":year})               #fianl dataframe
```
